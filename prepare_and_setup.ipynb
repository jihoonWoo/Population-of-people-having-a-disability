# Prepare and setup insight

Read data generated by "ABS-Health-and-Disability-2015" (https://zenodo.org/badge/latestdoi/361372177) to prepare and setup data insight.

## Download data

1. Go to (https://zenodo.org/badge/latestdoi/361372177) and download the repository. 
2. Unzip the repository on your working directory

## Import libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

## Read data

```python
input_path = '../../DATA/ABS-Health-and-Disability-2015/generated_data/'
input_file = 'transformed_data.csv'
```

```python
df = pd.read_csv(input_path + input_file, sep=',', encoding='utf-8')
df
```

The transformed data contains percentage of people having a disability by national and state level in 2015.

The data originally collected by Australian Bureau of Statistics (ABS) that is available at [2].

## Visualize data

Here, we'd like to visualize the generated data. Below code demonstrates how to visualize the data by using Matplot library.

**Matplot** : Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python.

```python

plt.style.use('ggplot')

x = list(df['Label'])
values = list(df['Percentage of people having a disability in 2015 (%)'])

x_pos = [i for i, _ in enumerate(x)]

plt.figure(figsize=(18,12))

plt.bar(x_pos, values, color='purple')

plt.xlabel("Labels", fontsize=22)
plt.ylabel("Percentage (%)", fontsize=22)
plt.title("Percentage of people having a disability in 2015 (%)", fontsize=28)

for index, value in enumerate(values):
    plt.text(index-0.2, value - 1, str(round(value, 2)), fontsize=18, color='white')

plt.xticks(x_pos, x, fontsize=18)

plt.savefig('img/Percentage of people having a disability in 2015 (%).pdf', bbox_inches='tight')

plt.show()
```

The above figure shows percentage of people having a disability in Australia in 2015. In Australia, there are 17.2 percent of people reported having a disability in 2015. And when we look into state level, TAS has the highest percentage (25.1%) followed by SA (21.8%) while NT has the lowest percentage (9.1%).

## Insight

**Around 17% of Australians reported having a disability in 2015.**

```python
value = float(df['Percentage of people having a disability in 2015 (%)'].iloc[0])
print('Around {}% of Australians reported having a disability in 2015.'.format(round(value, 0)))
```

## Conclusion

We extract insight "Around 17.0% of Australians reported having a disability in 2015." from the transformed data. The transformed data generated by [1] that it contains information about population of disability in Australia in 2015. According to the data, around 17% of Australians reported having a disability in 2015 where TAS and SA have higher percentage of people with disability than average.
